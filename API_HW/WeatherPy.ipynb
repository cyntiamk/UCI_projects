```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy
```

## Generate Cities List

```python
# empty lists to store new variables found
lats_lons = []
cities = []

# latitude range is from -90 to 90
# longitude range is from -180 to 180
# select random number between latitude and longitude
lats = np.random.uniform(-90,90,1500)
lons = np.random.uniform(-180,180,1500)

# zip the two lists together to get random coordinates
lats_lons = zip(lats,lons)

# using random coordinates stored in lats_lons list,
# search citipy for city names using those random coordinates
for lat_lon in lats_lons:
    city_coord = citipy.nearest_city(lat_lon[0], lat_lon[1])
    city = city_coord.city_name
    # only append city name that is unique
    if city not in cities:
        cities.append(city)
        
# convert the cities list into a dataframe naming the column City        
cities_df = pd.DataFrame(cities)
# add columns header
cities_df.columns = ["City"]
# save dataframe as csv
cities_df.to_csv("cities.csv",index=True, header=True)

# group cities_df by city to be able to add city number using cumcount
cities_grouped =cities_df.groupby("City")

# using cumcount, create a new column starting at 1 instead of 0
cities_df["City Number"] = np.arange(1, len(cities_grouped.cumcount()) + 1)
# cities_df.head()
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# using the variables extracted from the query url,
# we built a dictionary, and from that dictionary
# we built a dataframe and saved as csv
weather_dict = {"City":names,
                "Country":country_code,
                "Latitude":lat,
                "Longitude":lon,
                "Temperature(F)":temp,
                "Humidity(%)":hum,
                "Cloudiness(%)": cloud, 
                "Wind Speed(MPH)":wind,
                "Date":date}
weather_df = pd.DataFrame(weather_dict)
# converted the time in seconds to a more readable format
weather_df["Date"] = pd.to_datetime(weather_df["Date"],unit="s")

weather_df.to_csv("weather.csv")
weather_df.head()
```
